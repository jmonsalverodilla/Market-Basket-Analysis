## Funciones necesarias

```python
def descripcion(df:pd.DataFrame) -> None:
    '''
    df: pandas DataFrame que queremos describir
    Funcion que sirve para entender el número de nulos presentes en un df
    '''
    df_info = pd.DataFrame(df.dtypes).T.rename(index={0:'column type'})\
    .append(pd.DataFrame(df.count()).T.rename(index={0:'count'}))\
    .append(pd.DataFrame(df.nunique()).T.rename(index={0:'unique values'}))\
    .append(pd.DataFrame(df.isnull().sum()).T.rename(index={0:'null values'}))\
    .append(pd.DataFrame(df.isnull().sum()/df.shape[0]*100).T.rename(index={0:'null values (%)'}))
    display(df_info.head())
```

```python
def display_side_by_side(dfs:list, captions:list) -> None:
    """Display tables side by side to save vertical space
    Input:
        dfs: list of pandas.DataFrame
        captions: list of table captions
    """
    output = ""
    combined = dict(zip(captions, dfs))
    for caption, df in combined.items():
        output += df.style.set_table_attributes("style='display:inline'").set_caption(caption)._repr_html_()
        output += "\xa0\xa0\xa0"
    display(HTML(output))
```

## EDA

```python
# 1. Gather All Items of Each Transactions into Numpy Array
transaction = []
for i in range(0, df.shape[0]):
    for j in range(0, df.shape[1]):
        transaction.append(df.values[i,j])

# 2. Transform Them a Pandas DataFrame
df_items = pd.DataFrame(transaction, columns=["item"]) 
df_items["count"] = 1 # Put 1 to Each Item For Making Countable Table, to be able to perform Group By

# 3. Delete nan values from items column
df_items = df_items[~df_items['item'].isna()]

# 4. Groupy by 
df_items_grouped = df_items.groupby('item').agg({'count':'sum'}).reset_index().sort_values(by="count",ascending=False)
display(df_items_grouped)
px.bar(df_items_grouped.head(100), x="item",y="count").show()

# 5. Initial Visualizations
display(df_items_grouped.head(10).style.background_gradient(cmap='Blues'))

#6. Most demanded items
df_items_grouped["all"] = "all" # to have a same origin
fig = px.treemap(df_items_grouped.head(30), path=['all', "item"], values='count',
                  color=df_items_grouped["count"].head(30), hover_data=['item'],
                  color_continuous_scale='Blues',
                  )
fig.show()
```

## Customer's first choices

```python
# 1. Gather Only First Choice of Each Transactions into Numpy Array
# Similar Pattern to Above, Only Change is the Column Number "0" in Append Function
transaction = []
for i in range(0, df.shape[0]):
    transaction.append(df.values[i,0])

transaction = np.array(transaction)

# 2. Transform Them a Pandas DataFrame
df_first = pd.DataFrame(transaction, columns=["items"])
df_first["incident_count"] = 1

# 3. Delete NaN Items from Dataset
indexNames = df_first[df_first['items'] == "nan" ].index
df_first.drop(indexNames , inplace=True)

# 4. Final Step: Make a New Appropriate Pandas DataFrame for Visualizations  
df_table_first = df_first.groupby("items").sum().sort_values("incident_count", ascending=False).reset_index()
df_table_first["food"] = "food"
df_table_first = df_table_first.truncate(before=-1, after=15) # Fist 15 Choice
display(df_table_first.head(2))

# 5. Graph
plt.rcParams['figure.figsize'] = (9, 9)

first_choice = nx.from_pandas_edgelist(df_table_first, source = 'food', target = "items", edge_attr = True)
pos = nx.spring_layout(first_choice,seed=10)

nx.draw_networkx_nodes(first_choice, pos, node_size = 2000, node_color = "lavender")
nx.draw_networkx_edges(first_choice, pos, width = 3, alpha = 0.6, edge_color = 'black')
nx.draw_networkx_labels(first_choice, pos, font_size = 8, font_family = 'sans-serif')

plt.axis('off')
plt.grid()
plt.title('Top 15 First Choices', fontsize = 15)
plt.show()
```

## Data preprocessing

In order to be able to use apriori algorithm and get most frequent itemsets, we have to transform our dataset into a 1-0 matrix where rows are transactions and columns are products. On that matrix, "1" represents that a product has been bought on a certain transaction and "0" represents that the product has not been bought.

```python
transaction = []
for i in range(df.shape[0]):
    transaction.append([str(df.values[i,j]) for j in range(df.shape[1])])
#transaction = np.array(transaction)    

#Encoding to transform the matrix to false-true
te = TransactionEncoder()
te_ary = te.fit(transaction).transform(transaction)
dataset = pd.DataFrame(te_ary, columns=te.columns_)

#Transform to 1-0 matrix
def encode_units(x):
    if x == False:
        return 0 
    if x == True:
        return 1
    
dataset = dataset.applymap(encode_units)
dataset = dataset.drop(columns=['nan'])
dataset
```

## Algorithm implementation

### Support

*Support is an indication of how frequently the itemset appears in the dataset.*<br>
*In other words, this is an indication of how popular an itemset is in a dataset*
 ![Support](https://user-images.githubusercontent.com/36535914/81746687-98163000-94af-11ea-834a-8d9577f28930.png)

### Confidence

*Confidence is an indication of how often the rule has been found to be true* <br>
*In other words, confidence says how likely item Y is purchased when item X is purchased*

![Confidence](https://user-images.githubusercontent.com/36535914/82349639-094f6900-9a03-11ea-8163-8f4d9de06e14.png)

### Lift

*Lift is a metric to measure the ratio of X and Y occur together to X and Y occurrence if they were statistically independent. In other words, lift illustrates how likely item Y is purchased when item X is purchased, while controlling for how popular item Y is.*

* A Lift score that is close to 1 indicates that the antecedent and the consequent are independent and occurrence of antecedent has no impact on occurrence of consequent.

* A Lift score that is bigger than 1 indicates that the antecedent and consequent are dependent to each other, and the occurrence of antecedent has a positive impact on occurrence of consequent.

* A Lift score that is smaller than 1 indicates that the antecedent and the consequent are substitute each other that means the existence of antecedent has a negative impact to consequent or visa versa.

***Therefore having lift bigger than 1 is critial for proving associations***

![Lift](https://user-images.githubusercontent.com/36535914/85957011-5de6ec00-b992-11ea-8c5e-243f449d0241.png)

### Conviction

*Conviction measures the implication strength of the rule from statistical independence
Conviction score is a ratio between the probability that X occurs without Y while they were dependent and the actual probability of X existence without Y. For instance; if (French fries) --> (beer) association has a conviction score of 1.8; the rule would be incorrect 1.8 times as often (80% more often) if the association between totally independent.*

![Conviction](https://user-images.githubusercontent.com/36535914/82484880-42f7a100-9ae3-11ea-8219-cd27a9d96d6f.png)

### 5.1.5. Consequents & Antecedents

*The IF component of an association rule is known as the antecedent. The THEN component is known as the consequent. The antecedent and the consequent are disjoint; they have no items in common.*

![Cons_Ant](https://user-images.githubusercontent.com/36535914/82491863-efd71b80-9aed-11ea-80d5-2ffebd16251d.png)

```python
%%time
# Extracting the most frequest itemsets via Mlxtend.
# The length column has been added to increase ease of filtering.
#Let's stablish a minimum threshold of selecting those itemsets that appear at leat 2 times every 1000 transactions

frequent_itemsets = apriori(dataset, min_support=0.002, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
# Association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
rules["antecedents_length"] = rules["antecedents"].apply(lambda x: len(x))
rules["consequents_length"] = rules["consequents"].apply(lambda x: len(x))
display(rules.shape[0])
rules.sort_values("confidence",ascending=False)
```

According to above table, the customers who bought **(mushroom cream sauce, pasta)** are expected to buy **(escalope)** with a likelihood of 95% (confidence). Lift & conviction scores support that hypothesis too
It would be better to keep them close to increase sales !

## Let's see what the algorithm predicts as next product to buy

```python
items = ['low fat yogurt','cottage cheese','pancakes','brownies']
names = list(range(0, len(items)))
number_of_predictions = 100

dict_predictions = {}
for item,name in zip(items,names):
    df_prediction = rules[(rules["antecedents"].str.contains(item, regex=False)) & (rules["antecedents_length"] == 1)].sort_values("confidence", ascending=False)
    dict_predictions[name] = df_prediction
df_all_predictions = pd.concat(dict_predictions[name] for name in dict_predictions.keys())
df_all_predictions = df_all_predictions[df_all_predictions['consequents_length']==1]
df_all_predictions["antecedents"] = df_all_predictions["antecedents"].apply(lambda x: list(x)[0]).astype("unicode")
df_all_predictions["consequents"] = df_all_predictions["consequents"].apply(lambda x: list(x)[0]).astype("unicode")
df_all_predictions = df_all_predictions[~df_all_predictions["consequents"].isin(items)]
display(df_all_predictions[df_all_predictions['consequents']=='spaghetti'])

#Let's get multiplication, sum and average of confidence
df_result_1 = df_all_predictions.groupby(["consequents","consequent support"]).agg({'confidence':['sum','mean']}).reset_index()
df_result_1.columns = ['consequents','consequent support','confidence_sum','confidence_mean']

df_result_2 = pd.DataFrame(df_all_predictions.groupby('consequents')['confidence'].apply(lambda x : x.cumprod().iloc[-1])).reset_index().rename(columns={'confidence':'confidence_multiply'})

df_result = df_result_1.merge(df_result_2,how="inner",on="consequents")

df_sum = df_result.sort_values(by="confidence_sum",ascending=False).head(number_of_predictions)
df_mean = df_result.sort_values(by="confidence_mean",ascending=False).head(number_of_predictions)
df_multiply = df_result.sort_values(by="confidence_multiply",ascending=False).head(number_of_predictions)

display_side_by_side([df_sum],['df_sum'])
```
